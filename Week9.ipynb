# 10 Mengenal Text Processing: Bag of Words & Stop Word Filtering

## Bag of Words model sebagai representasi text

### Dataset

```python
corpus = [
    'Linux has been around since the mid-1990s.',
    'Linux distributions include the Linux kernel.',
    'Linux is one of the most prominent open-source software.'
]

corpus
```

### Bag of Words model dengan CountVectorizer

```python
# from sklearn.feature_extraction.text import CountVectorizer

# vectorizer = CountVectorizer()
# vectorized_X = vectorizer.fit_transform(corpus).todense()
# vectorized_X
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized_X = vectorizer.fit_transform(corpus).toarray()
vectorized_X
```

```python
# vectorizer.get_feature_names()
```

```python
vectorizer.get_feature_names_out()
```

### Euclidean Distance untuk mengukur kedekatan/jarak antar dokumen (vector)

```python
# from sklearn.metrics.pairwise import euclidean_distances

# for i in range(len(vectorized_X)):
#     for j in range(i, len(vectorized_X)):
#         if i == j:
#             continue
#         jarak = euclidean_distances(vectorized_X[i], vectorized_X[j])
#         print(f'Jarak dokumen {i+1} dan {j+1}: {jarak}')
```

```python
from sklearn.metrics.pairwise import euclidean_distances

distances = euclidean_distances(vectorized_X)

for i in range(len(distances)):
    for j in range(i, len(distances)):
        if i == j:
            continue
        jarak = distances[i][j]
        print(f'Jarak dokumen {i+1} dan {j+1}: {jarak}')
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')
vectorized_X = vectorizer.fit_transform(corpus).toarray()
vectorized_X
```

# 11 Mengenal TF-IDF (Term Frequency - Inverse Document Frequency)

## Dataset

```python
corpus = [
    'the house had a tiny little mouse', 
    'the cat saw the mouse',
    'the mouse ran away from the house', 
    'the cat finally ate the mouse',
    'the end of the mouse story'
]

corpus
```

## TF-IDF Weights dengan TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
response = vectorizer.fit_transform(corpus)
print(response)
```

```python
vectorizer.get_feature_names_out()
```

```python
response.todense()
```

```python
import pandas as pd

df = pd.DataFrame(response.todense().T,
                 index=vectorizer.get_feature_names(),
                 columns=[f'D{i+1}' for i in range(len(corpus))])

df
```
